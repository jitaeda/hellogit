```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 18
plt.rcParams["figure.figsize"] = (20, 10)
plt.rcParams["axes.unicode_minus"] = False
```

```python
# 2017년 1월~12월 도시별 교통사고 건수
seoul = [3166, 2728, 3098, 3172, 3284, 3247, 3268, 3308, 3488, 3312, 3375, 3179]
busan = [927, 857, 988, 955, 1014, 974, 1029, 1040, 1058, 971, 958, 982]
daegu = [933, 982, 1049, 1032, 1083, 1117, 1076, 1080, 1174, 1163, 1146, 1135]
inchun = [655, 586, 629, 669, 643, 627, 681, 657, 662, 606, 641, 663]
label = ['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']

fig, cx = plt.subplots(1, 1)
cx.grid()
cx.set_title('2017년 주요도시 교통사고')
cx.set_xlabel('월')
cx.set_ylabel('교통사고')

cx.plot(seoul, label='서울')
cx.plot(busan, label='부산')
cx.plot(daegu, label='대구')
cx.plot(inchun, label='인천')
plt.savefig('traffic1.png', dpi=200, bbox_inches='tight')

cx.set_xlim([0,11])
cx.set_ylim([0,4000])

cx.legend(title='도시', loc='center left')
plt.savefig('traffic2.png', dpi=200, bbox_inches='tight')

x = list(range(0, len(label)))

cx.set_xticks(x)
cx.set_xticklabels(label)

plt.savefig('traffic3.png', dpi=200, bbox_inches='tight')

plt.show()
plt.close()
```

# 막대그래프

```python
#샘플데이터
cnt_dict = {'아이언맨':59, '캡틴아메리카':23, '호크아이':5, '토르':38, 
            '헐크':14, '블랙위도우':16}

values = list(cnt_dict.values())
print(values)
```

```python
keys = list(cnt_dict.keys())
print(keys)
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'

#그래프 시작
fig, ax = plt.subplots(1, 1, figsize = (14,7))

ax.bar(keys, values, label='득표수')
ax.legend()
ax.set_xlabel('케릭터', color='#ff0000', fontsize=14)
ax.set_ylabel('득표수', color='#0000ff', fontsize=14)
ax.set_title('어벤져스 인기투표', color='#ff6600', fontsize=24, pad=16)
ax.grid()
for i, v in enumerate(values):
    str_val = "%d표" % v
    ax.text(keys[i], v, str_val, fontsize=13, color='#0066ff', 
           horizontalalignment='center', verticalalignment='bottom')
    
plt.savefig('avengers2.png', dpi=200, bbox_inches='tight')

plt.show()
plt.close()
```

```python
newborn = [436455, 435435, 438420, 406243, 357771]

x = list(range(0,len(newborn)))
print(x)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(14,7))

ax.bar(x, newborn, label='신생아 수', color='#ecc2f9')
ax.legend()
ax.set_xlabel('년도')
ax.set_ylabel('신생아 수')
ax.set_title('2013~2017년도 신생아 수', fontsize=24, color='#0066ff', pad=20)
ax.grid()
ax.set_xticks(x)
ax.set_xticklabels(['2013년', '2014년', '2015년', '2016년', '2017년'])

for i,v in enumerate(newborn):
    str_val = "%d명" % v
    ax.text(x[i], v, str_val, fontsize=20, color='#ff0000', horizontalalignment = 'center',
           verticalalignment = 'bottom')
    
ax.set_ylim([300000, 450000])

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(18,9))

ax.barh(year, newborn, label='신생아 수')
ax.legend()
ax.set_ylabel('년도')
ax.set_xlabel('신생아 수')
ax.set_xlim([350000,450000])
ax.set_title('년도별 신생아 수')
ax.grid()

for i, v in enumerate(year):
    str_val = "%s명" % format(newborn[i], ",")
    ax.text(newborn[i], v, str_val, fontsize = 16, color='#ff6600',
           horizontalalignment = 'left', verticalalignment = 'center')

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(1,1,figsize=(18,9))
ax.barh(x-0.2, 중간고사, label='중간고사', height=0.4, color='#f2e3f7')
ax.barh(x+0.2, 기말고사, label='기말고사', height=0.4, color='#ecc2f9')

ax.set_yticks(x)
ax.set_yticklabels(과목명)

ax.legend()
ax.grid()
ax.set_ylabel('과목', fontsize=16, color='#0066ff')
ax.set_xlabel('점수', fontsize=16, color='#0066ff')
ax.set_xlim([0,110])
ax.set_title('철수의 이번학기 과목별 성적', fontsize=28, pad=14)

plt.show()
plt.close()
```

```python
ratio = [3700,2800,1300,590,600]
labels = ['Apple', '삼성', 'LG', 'Sony', '기타']
colors = ['#ff0000','#0066ff','#ff00ff','#00ffff','#ffff00']
explode = [0.0, 0.0, 0.0, 0.2, 0.1]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.set_title('스마트폰 점유율', fontsize=24)

ax.pie(ratio, labels=labels, colors=colors, explode=explode,
      autopct='%0.2f%%', shadow=False, startangle=-60)

plt.savefig("pie.png", dpi=200, bbox_inches='tight')
plt.show()
plt.close()
```
