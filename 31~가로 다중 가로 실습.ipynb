# 가로막대 그래프 

```python
from matplotlib import pyplot
from sample import newborn
from sample import year

pyplot.rcParams["font.family"] = 'NanumGothic'
pyplot.rcParams["font.size"] = 12
pyplot.rcParams["figure.figsize"] = (12, 8) #그대로

pyplot.figure()

pyplot.barh(year, newborn, label = '신생아 수', height = 0.7) #y, x 반대로
pyplot.legend()
pyplot.ylabel('년도')
pyplot.xlabel('신생아 수')
pyplot.xlim(350000, 450000)
pyplot.title('년도별 신생아 수')
pyplot.grid

for i, v in enumerate(year):
    str_val = "%d명" % newborn[i]
    pyplot.text(newborn[i], v, str_val, fontsize = 10, color = '#ff6600',
               horizontalalignment = 'left', verticalalignment = 'center') #중심축 확인

pyplot.show()
pyplot.close()
```

# 다중 가로 막대 그래프

```python
import numpy
from matplotlib import pyplot
from sample import seoul
from sample import busan
from sample import label

y = numpy.arange(len(label))
#print(y)

pyplot.rcParams["font.family"] = 'NanumGothic'
pyplot.rcParams["font.size"] = 12
pyplot.rcParams["figure.figsize"] = (12, 8)

pyplot.figure()

pyplot.barh(y+0.2, seoul, label = '서울', height = 0.4, color = '#ff0000')
pyplot.barh(y-0.2, busan, label = '부산', height = 0.4, color = '#0066ff')

pyplot.yticks(y, label) # y축 라벨 별도로 지정함

pyplot.legend()
pyplot.ylabel('월')
pyplot.xlabel('교통사고 수')
pyplot.xlim(0,4000)
pyplot.title('2017년도 서울, 부산 교통사고 현황')

pyplot.show()
pyplot.close()
```
