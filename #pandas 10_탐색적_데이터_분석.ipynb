```python
import pandas as pd
from pandas import DataFrame
from pandas import read_excel
from pandas import merge
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('http://itpaper.co.kr/data/exam_eda_12.txt', delimiter = '  ', header = None,\
                encoding = 'euc_kr', engine = 'python')
df
```

```python
df0 = df.dropna(axis=1)
df0.columns = ['a','b','c','d','e']
df0
```

```python
dfs = df0.melt()
df1 = dfs.drop(dfs.columns[0], axis = 1)
df1.columns = ['수명']
df1
```

## 26-(2)번

```python
df1.describe()
```

```python
def x(x):
    return x - 24.088
df2 = df1.agg([x])
df2
```

```python
df3 = df2['수명']
df3
```

```python
df4 = pd.concat([df1,df3], axis = 1)
df4
```

## 26-(3)번

```python
count = len(df4['수명'])
print(count)
step = 5
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 16
plt.rcParams['figure.figsize'] = (7, 5)
plt.rcParams['axes.unicode_minus'] = False
```

```python
fig, ax = plt.subplots(1,2, figsize = (15,8))

ax[0].hist(df4['수명'], bins = step)
ax[0].grid()
ax[1].boxplot(df4['수명'])

plt.show()
plt.close()
```
