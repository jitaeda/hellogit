```python
from matplotlib import pyplot
from sample import vote
from collections import Counter
cnt = Counter(vote)
print(cnt)
```

```python
cnt_dict = dict(cnt)
print(cnt_dict)
```

```python
values = list(cnt_dict.values())
print(values)
```

```python
keys = list(cnt_dict.keys())
print(keys)
```

```python
pyplot.rcParams["font.family"] = 'NanumGothic'
pyplot.rcParams["font.size"] = 12
pyplot.rcParams["figure.figsize"] = (12,8)

pyplot.figure()
pyplot.bar(keys, values, label = '득표수')
pyplot.legend()
pyplot.xlabel('케릭터')
pyplot.ylabel('득표수')
pyplot.title('어벤저스 인기투표')
pyplot.grid()
pyplot.savefig('avengers1.png')

for i, v in enumerate(values):
    str_val = "%d표" % v
    pyplot.text(keys[i], v, str_val, fontsize=9, color='#ff6600', 
               horizontalalignment='center', verticalalignment='bottom')
pyplot.savefig('avengers2.png')

pyplot.show()
pyplot.close()
```

```python
from matplotlib import pyplot
from sample import newborn
x = list(range(0, len(newborn)))
print(x)
```

```python
pyplot.rcParams["font.family"] = 'NanumGothic'
pyplot.rcParams["font.size"] = 12
pyplot.rcParams["figure.figsize"] = (12, 8)

pyplot.figure()
pyplot.bar(x, newborn, label='신생아 수', color=['#ecc2f9', '#ccffff', '#ff6600', '#ff0000', '#66ff00'])#개별 색상 설정
                                                                       #막대의 수만큼 색상값을 리스트로 설정
pyplot.legend()
pyplot.xlabel('년도')
pyplot.ylabel('신생아 수')
pyplot.title('2013~2017년도 신생아 수')
pyplot.grid()

pyplot.savefig('newborn1.png', dpi=200) #x가 0~4 리스트로 표시 -> 년도로 바꾸기

pyplot.xticks(x, ['2013년', '2014년', '2015년', '2016년', '2017년'])
pyplot.savefig('newborn2.png', dpi=200)

for i, v in enumerate(newborn):
    str_val = "%d명" % v
    pyplot.text(x[i], v, str_val, fontsize = 9, color = '#ff0000',
               horizontalalignment = 'center', verticalalignment = 'bottom')

    pyplot.savefig('newborn3.png', dpi=200)
    
pyplot.ylim(300000, 450000)  #아래를 잘라서 강조효과
pyplot.savefig('newborn4.png', dpi=200)

pyplot.show()
pyplot.close()
```
