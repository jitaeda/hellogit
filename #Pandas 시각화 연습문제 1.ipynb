# 문제 1

```python
from pandas import DataFrame
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
import numpy as np
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 16
plt.rcParams['figure.figsize'] = (7, 5)
plt.rcParams['axes.unicode_minus'] = False
```

```python
source1 = [
    [20, "10.0%"],
    [55, "27.5%"],
    [68, "34.0%"],
    [57, "28.5%"]
]

df1 = DataFrame(source1, index = ["10대", "20대", "30대", "40대"],
                columns = ["사람수", "비율"])
df1
```

```python
age_list = []
for  i in df1.index:
    age_list.append(i)
print(age_list)

x = list(range(0, len(age_list)))
print(x)
```

```python
fig, ax = plt.subplots(1, 1)

df1['사람수'].plot.bar(color = '#ff0000', rot = 0, width = 0.2, ax=ax)

ax.grid()
ax.legend()
ax.set_title("XX전자상가 고객 연령별 분포", fontsize = 30, pad = 20)
ax.set_xlabel("연령군", fontsize = 20)
ax.set_ylabel("사람수", fontsize = 20)

ax.set_xticks(x)
ax.set_xticklabels(age_list)
ax.set_ylim([0, 80])

for i, y in enumerate(df1['사람수']):
    txt = "%s건" % format(y, ",")
    ax.text(x[i], y, txt, fontsize=14, color = '#8046eb',
           horizontalalignment = 'center', verticalalignment = 'bottom')

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (15, 10))
df1['사람수'].plot.pie(labels = age_list,
                    explode = [0.05, 0.03, 0.00, 0.02],
                    autopct='%0.1f%%',
                    startangle = -167,
                    ax=ax)

ax.set_title("XX전자상가 고객 연령별 분포", fontsize = 28)
plt.show()
plt.close()
```

# 문제 3

```python
source3 = {
    '연도': [1990, 1995, 2000, 2005, 2010, 2012, 2013, 2030, 2040],
    '총인구': [42689, 45093, 47008, 48138, 49410, 50004, 50220, 52160, 51091],
    '인구성장률': [0.99, 1.01, 0.84, 0.21, 0.46, 0.45, 0.43, 0.01, -0.39]
}

dfc = DataFrame(source3)
df3 = dfc.set_index('연도')
df3
```

```python
xtext3 = []
for i in df3.index:
    xtext3.append("%s년" % i)
print(xtext3)

xpos3 = list(range(0, len(xtext3)))
print(xpos3)

tnum = list(df3['총인구'])
print(tnum)
pnum = list(df3['인구성장률'])
print(pnum)
```

```python
plt.rcParams['figure.figsize'] = (12, 10)

fig, ax1 = plt.subplots(1, 1)
ax2 = ax1.twinx()

sns.barplot(x=xtext3, y=tnum, label='총인구', palette='pastel', ax=ax1)
sns.lineplot(x=xtext3, y=pnum, label='인구성장률', 
             linestyle = '--', marker = '.', color = '#ff0000', ax=ax2)

ax2.legend(loc='upper center')
ax1.set_xlabel('연도', color = '#ff0000', fontsize = 14)
ax1.set_ylabel('총인구', color = '#ff6600', fontsize = 14)
ax1.set_title('우리나라의 총 인구와 인구성장률 추이', color = '#0000ff', fontsize = 20, pad = 16)
ax1.grid()
ax1.set_ylim(30000, 55000)
ax2.set_ylim(-1.2, 1.2)

plt.show()
plt.close()
```

# 문제 5
### 링크 http://itpaper.co.kr/data/ad-sales.xlsx

```python
adxlsx = read_excel("http://itpaper.co.kr/data/ad-sales.xlsx", engine = 
                    "openpyxl")
adxlsx
ad5 = adxlsx.set_index('월')
ad5
```

```python
month = []
for i in ad5.index:
    month.append("%s" % i)
print(month)

xpos5 = list(range(0, len(month)))
print(xpos5)

adp = list(ad5['광고비(백만원)'])
print(adp)
sell = list(ad5['매출액(억원)'])
print(sell)
```

### (1)

```python
plt.rcParams['figure.figsize'] = (12, 10)

fig, ax1 = plt.subplots(1, 1)
ax2 = ax1.twinx()

sns.barplot(x=month, y=adp, label='광고비', ax=ax1) #data=ad5 하면 x, y 리스트로 안 빼도 됨
sns.lineplot(x=month, y=sell, label='매출액', #data=ad5 하면 x, y 리스트로 안 빼도 됨
             linestyle = '-', marker = '.', color = '#ff0000', ax=ax2)
ax1.legend(loc='upper center')
ax2.legend()
ax1.set_xlabel('연도', color = '#ff0000', fontsize = 14)
ax1.set_ylabel('광고비(백만원)', color = '#ff6600', fontsize = 14)
ax2.set_ylabel('매출액(억원)', color = '#ff6600', fontsize = 14)
ax1.set_title('광고비와 매출액', color = '#0000ff', fontsize = 20, pad = 16)
ax1.grid()
ax1.set_ylim(0, 270)
ax2.set_ylim(0, 950)


plt.show()
plt.close()
```

### (2)

```python
mon2 = []
for i in month:
    mon2.append('%s' % i)
    mon2.append('%s' % i)
print(mon2)

matter = list(ad5.columns)*12
print(matter)

sumoney = []
for i in range(0,len(adp)):
    sumoney.append(adp[i])
    sumoney.append(sell[i])        
print(sumoney)
```

```python
fig, ax = plt.subplots(1, 1)

sns.barplot(x=mon2, hue=matter, y=sumoney, palette = "dark", ax=ax)
ax.set_title("광고비용 대비 매출액")
ax.grid()
ax.legend()
plt.show()
plt.close()
```

### (3)

```python
fig, ax = plt.subplots(1, 2, figsize = (35,10))

ax[0].bar(month, adp)
ax[1].bar(month, sell)

plt.show()
plt.close()
```
